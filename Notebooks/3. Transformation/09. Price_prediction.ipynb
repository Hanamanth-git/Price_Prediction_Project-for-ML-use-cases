```python
%sql
select 
daily_pricing.DATE_OF_PRICING,
daily_pricing.STATE_NAME,
daily_pricing.MARKET_NAME,
daily_pricing.PRODUCT_NAME AS PRODUCT_ID,
daily_pricing.PRODUCTGROUP_NAME as PRODUCT_GROUP_ID,
daily_pricing.ROW_ID,
daily_pricing.ARRIVAL_IN_TONNES,
daily_pricing.MINIMUM_PRICE,
daily_pricing.MAXIMUM_PRICE,
daily_pricing.MODAL_PRICE,

geo_location.latitude as MARKET_LATITUDE,
geo_location.longitude as MARKET_LONGITUDE,
geo_location.population as MARKET_POPULATION,

weather.unitOfTemparature as TEMPERATURE_UNIT,
weather.maximumTemparature as MARKET_MAX_TEMPARATURE,
weather.minimumTemparature as MARKET_MIN_TEMPARATURE,
weather.unitOfRainFall as RAINFALL_UNIT,
weather.rainFall as MARKET_DAILY_RAINFALL,
current_timestamp() as lakehouse_inserted_date,
current_timestamp() as lakehouse_updated_date

from pricing_analytics.silver.daily_pricing daily_pricing
join pricing_analytics.silver.geolocation_date geo_location
on daily_pricing.STATE_NAME=geo_location.STATE_NAME and daily_pricing.MARKET_NAME=geo_location.MARKET_NAME and geo_location.COUNTRY='India'
join pricing_analytics.silver.weather_date weather
on daily_pricing.MARKET_NAME=weather.MARKET_NAME and daily_pricing.DATE_OF_PRICING=weather.weatherDate    
```
