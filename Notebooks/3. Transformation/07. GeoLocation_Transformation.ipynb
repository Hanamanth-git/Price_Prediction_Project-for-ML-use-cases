```python
geo_location_df=spark.table("pricing_analytics.bronze.geo_location_data")
```

```python
from pyspark.sql.functions import explode, col, current_timestamp
geo_location_df=geo_location_df.withColumn('results', explode('results'))
```

```python
df=geo_location_df.select(col('results.admin1').alias('STATE_NAME'), col('results.admin2').alias('DISTRICT_NAME'), col('results.country').alias('COUNTRY'), col('results.latitude').alias('latitude'), col('results.longitude').alias('longitude'), col('results.name').alias('MARKET_NAME'), col('results.population').alias('population')).withColumn('lake_house_inserted_date', current_timestamp()).withColumn('lake_house_updated_date', current_timestamp())
```

```python
df.write.format('delta').mode("append").saveAsTable('pricing_analytics.silver.Geolocation_date')
```
